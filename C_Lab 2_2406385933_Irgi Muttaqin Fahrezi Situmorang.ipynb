# Exploratory Data Analysis
**Exploratory Data Analysis (EDA)** adalah proses untuk mengeksplorasi dan memahami pola serta karakteristik utama dari suatu dataset. EDA merupakan tahap awal yang krusial dalam analisis data karena hasil dari EDA dapat mempengaruhi keputusan terkait data processing, data modeling, dan langkah-langkah analitis selanjutnya.

Proses EDA melibatkan dua komponen utama: **statistik deskriptif** dan **visualisasi data**. Melalui statistik deskriptif, kita menghitung metrik seperti mean, median, mode, dan standar deviasi untuk memahami distribusi dan variasi data. Sedangkan, melalui visualisasi data, kita dapat melihat pola dan tren dalam data menggunakan grafik seperti histogram, box plot, dan scatter plot. EDA juga dapat dibagi menjadi beberapa jenis analisis seperti **Univariate Analysis**, **Bivariate Analysis**, dan **Multivariate Analysis**.

EDA membantu kita mengidentifikasi masalah potensial dan mengarahkan analisis lebih lanjut untuk mendapatkan wawasan berharga dari data.

## Descriptive Statistics
Terdapat berbagai pengukuran yang dapat diterapkan untuk statistik deskriptif. Pengukuran ini sudah pernah Anda dapatkan dalam mata kuliah Statistika dan Probabilitas:
1. *Central tendency* (mean, median, mode)
2. *Dispersion* atau *variablity* (variance, standard deviation)
3. Minimum, maximum
4. Range
5. Quartile
6. Percentile

**Central tendency data dapat digunakan untuk memahami distribusi data.** `Mean` biasanya lebih tepat digunakan untuk data dengan distribusi normal (di mana data tersebar secara simetris di sekitar pusat). Di sisi lain, `median` lebih sesuai untuk distribusi yang miring (*skewed*), di mana data condong ke satu sisi. Untuk data kategorikal, hanya `mode` yang dapat digunakan.

Namun, *central tendency* saja tidak cukup untuk menggambarkan distribusi data secara keseluruhan, karena tidak memberikan informasi mengenai tingkat penyebaran data. **Measure of dispersion** adalah alat ukur yang digunakan untuk memahami **variasi atau seberapa luas data tersebar**. Beberapa cara untuk mengukur dispersi data termasuk menggunakan `range` (jangkauan) dan `standard deviation` (simpangan baku).

Berikut adalah penggunaan statistik deskriptif (1), (2), (3) dari atribut `TotalCharges`:

```python
total_charges_arr = df["TotalCharges"].dropna()
stats = [
    ['Mean', np.mean(total_charges_arr)],
    ['Median', np.median(total_charges_arr)],
    ['Mode', scp.mode(total_charges_arr)[0]],
    ['Standard Deviation', np.std(total_charges_arr)],
    ['Variance', np.var(total_charges_arr)],
    ['Minimum', np.min(total_charges_arr)],
    ['Maximum', np.max(total_charges_arr)],
]
pd.DataFrame(stats, columns=['Measure', 'Value'])
```

Pada hasil di atas, bisa kita lihat bahwa nilai terbesar dimiliki oleh Mean dengan urutan sebagai berikut: Mean > Median > Mode.

Data tersebar sebesar 515.38 dari nilai Mean, menunjukkan bahwa ada variasi yang signifikan dalam data. Nilai maksimum adalah 2252.49 dan nilai minimum adalah 19.38. Ini menunjukkan bahwa data cenderung tersebar di berbagai rentang nilai dengan nilai yang lebih tinggi cenderung lebih banyak. Dengan kata lain, kemungkinan data ini padat di nilai yang tinggi atau padat ke kanan, mengindikasikan distribusi yang mungkin memiliki skewness positif.

```python
plt.figure(figsize=(10,5))
sns.kdeplot(total_charges_arr, shade=True)
plt.vlines(x=np.mean(total_charges_arr), ymin=0, ymax=0.001, color='blue', linestyle='--')
plt.vlines(x=np.median(total_charges_arr), ymin=0, ymax=0.001, color='brown', linestyle='--')
plt.vlines(x=scp.mode(total_charges_arr)[0], ymin=0, ymax=0.001, color='red', linestyle='--')
plt.legend(['Total Charges', 'Mean', 'Median', 'Mode'], fontsize=12)
plt.show()
```

Dari visualisasi distribusi di atas, bisa dilihat bahwa distribusi bersifat `right-skewed` karena distribusi data lebih berat di sebelah kiri dan memiliki ekor yang panjang di sebelah kanan. Ini berarti ada beberapa nilai yang jauh lebih besar dari sebagian besar data lainnya.

Dapat terlihat pula pada visualisasi data di atas bahwa `mean`, `median`, dan `mode` tidak sama, yang merupakan karakteristik umum dari distribusi skewed. Dalam kasus ini:

- `Mean` (ditandai dengan garis biru) lebih besar dari `Median` (ditandai dengan garis coklat) dan `Mode` (ditandai dengan garis merah), mengindikasikan bahwa beberapa nilai besar menarik rata-rata ke arah kanan.
- `Median` terletak di antara `Mean` dan `Mode`, memberikan indikasi yang lebih representatif tentang central tendency dari data dalam kasus distribusi skewed ini.

Pada distribusi yang skewed ke kanan, median sering kali menjadi pengukuran yang lebih baik untuk menggambarkan central tendency karena tidak terpengaruh oleh nilai ekstrem sebesar mean. Oleh karena itu, median memberikan pandangan yang lebih stabil terhadap data yang terdistribusi skewed.

# Korelasi

Korelasi dalam *data science* mengacu pada hubungan statistik antara dua atau lebih variabel. Dengan memanfaatkan korelasi, kita dapat memahami bagaimana perubahan pada satu variabel terkait dengan perubahan pada variabel lainnya. Korelasi mengukur sejauh mana dua variabel bergerak bersama-sama atau memiliki ketergantungan dalam pola tertentu.

Ada dua jenis korelasi yang sering digunakan:

* Korelasi Positif: Ketika dua variabel memiliki korelasi positif, ini berarti ketika satu variabel meningkat, variabel lainnya juga cenderung meningkat, dan sebaliknya. Hal ini menunjukkan adanya hubungan searah antara kedua variabel tersebut.

* Korelasi Negatif: Jika dua variabel memiliki korelasi negatif, artinya ketika satu variabel naik, variabel lainnya cenderung turun, dan sebaliknya. Ini menunjukkan adanya hubungan terbalik antara kedua variabel.

Korelasi dapat diukur menggunakan koefisien korelasi, dengan koefisien **korelasi Pearson** sebagai salah satu yang paling umum digunakan. Koefisien ini memiliki rentang nilai antara -1 hingga 1:

* Nilai +1 menunjukkan korelasi positif sempurna, artinya kedua variabel bergerak searah secara linier.
* Nilai -1 menunjukkan korelasi negatif sempurna, artinya kedua variabel bergerak berlawanan arah secara linier.
* Nilai 0 menunjukkan tidak ada korelasi linier antara kedua variabel tersebut.

Korelasi adalah alat penting dalam analisis data karena membantu kita mengidentifikasi hubungan antar variabel, memprediksi perilaku data atau variabel, dan membuat keputusan yang lebih baik berdasarkan pemahaman tentang bagaimana variabel-variabel tersebut saling berkaitan. Namun, perlu diingat bahwa korelasi hanya menunjukkan adanya hubungan atau pola antara variabel, bukan hubungan sebab-akibat atau kausalitas. Oleh karena itu, kehati-hatian sangat diperlukan dalam menginterpretasikan hasil korelasi.

```python
# Seleksi hanya kolom numerik untuk perhitungan korelasi Pearson
numeric_df = df.select_dtypes(include=[np.number])

# Implementasi korelasi Pearson hanya untuk kolom numerik
correlation_matrix = numeric_df.corr(method='pearson')
correlation_matrix
```

Berdasarkan hasil perhitungan korelasi Pearson di atas, kita dapat mengidentifikasi pasangan atribut yang memiliki relasi tinggi dengan atribut lainnya. Dalam kasus ini, beberapa pasangan atribut yang memiliki relasi kuat adalah:

1. **`AccountAge` dengan `TotalCharges`** memiliki korelasi positif yang sangat tinggi sebesar **0.84**. Hal ini menunjukkan bahwa semakin lama usia akun, semakin tinggi total biaya yang telah dikeluarkan.
2. **`MonthlyCharges` dengan `TotalCharges`** memiliki korelasi positif yang cukup kuat sebesar **0.50**. Ini menunjukkan bahwa terdapat hubungan antara besarnya biaya bulanan dan total biaya yang dikeluarkan.
3. **`Churn` dengan `AccountAge`** memiliki korelasi negatif sebesar **-0.22**. Hal ini menunjukkan bahwa semakin lama usia akun, semakin kecil kemungkinan pelanggan untuk melakukan churn.
4. **`Churn` dengan `TotalCharges`** juga menunjukkan korelasi negatif sebesar **-0.15**, yang berarti semakin tinggi total biaya yang dikeluarkan, semakin kecil kemungkinan pelanggan melakukan churn.
5. **`Churn` dengan `MonthlyCharges`** menunjukkan korelasi positif sebesar **0.13**. Ini menunjukkan bahwa semakin tinggi biaya bulanan, semakin besar kemungkinan pelanggan untuk melakukan churn.



Selain metode *pearson*, ada metode korelasi lain yang pernah dipelajari di mata kuliah Statistika dan Probabilitas, yaitu metode `spearman rho`. Perbedaan mendasar antara kedua metode tersebut adalah bahwa metode *pearson* dihitung berdasarkan perbedaan antara nilai sebenarnya dan rata-rata, kemudian dibagi dengan standar deviasi, sedangkan metode `spearman rho` dihitung berdasarkan peringkat (peringkat ordinal) dari nilai setiap variabel. Karena penghitungan dengan *pearson* menggunakan rata-rata dan varians, skala pengukuran atribut harus dipastikan dalam bentuk interval dan rasio. Sebaliknya, penghitungan `spearman rho` menggunakan peringkat, sehingga metode ini juga dapat mengukur korelasi untuk atribut dengan skala pengukuran ordinal.

```python
correlation_matrix = numeric_df.corr(method='spearman')
correlation_matrix
```

Jika dilihat dari perbandingan nilai pada tabel korelasi Pearson dan Spearman, korelasi antara AccountAge dan TotalCharges menunjukkan bahwa nilai Spearman sebesar 0.874 lebih tinggi daripada Pearson yang bernilai 0.835. Hal ini mengindikasikan bahwa meskipun hubungan antara AccountAge dan TotalCharges bersifat monotonic (keduanya bergerak searah), hubungan tersebut mungkin tidak sepenuhnya linier. Di sisi lain, untuk MonthlyCharges dan TotalCharges, korelasi Pearson (0.497) lebih tinggi daripada Spearman (0.453), yang menunjukkan bahwa hubungan linier lebih dominan.

Secara umum, ketika korelasi Spearman lebih tinggi dari Pearson, ini menandakan bahwa hubungan antara dua variabel bersifat monotonic tetapi mungkin tidak secara sempurna mengikuti garis lurus. Sebaliknya, jika korelasi Pearson lebih tinggi, ini menunjukkan bahwa hubungan linier lebih kuat, di mana perubahan pada satu variabel diikuti oleh perubahan proporsional pada variabel lainnya, dengan sedikit atau tanpa penyimpangan dari garis lurus.

Penjelasan ini menggarisbawahi bahwa hubungan linier juga merupakan hubungan monotonic, tetapi hubungan monotonic tidak selalu linier.

Kita akan membuat visualisasi korelasi antar atribut menggunakan scatterplot untuk lebih memahami hubungan di antara variabel-variabel tersebut.

```python
# Scatterplot AccountAge vs TotalCharges
df_subset = df[["AccountAge", "TotalCharges"]]
plt.figure(figsize=(8, 6))
plt.scatter(df_subset["AccountAge"], df_subset["TotalCharges"], alpha=0.5, label="Data Points")

# Perhitungan linear regression
w = df_subset["AccountAge"]
z = df_subset["TotalCharges"]
m, b = np.polyfit(w, z, 1)  # m = slope, b = intercept

# Plot regression line
plt.plot(w, m * w + b, color='red', label=f"Regression Line (y={m:.2f}x + {b:.2f})")
plt.xlabel("Account Age")
plt.ylabel("Total Charges")
plt.title("Scatter Plot of Account Age vs Total Charges")
plt.legend()
plt.show()
```

Scatter plot ini menunjukkan adanya hubungan positif yang signifikan antara `Account Age` dan `Total Charges`, di mana peningkatan usia akun cenderung berkorelasi dengan peningkatan total biaya yang terkait dengan akun tersebut. Garis regresi linier yang dihasilkan, dengan persamaan \( y = 13.17x - 27.62 \), mengindikasikan bahwa untuk setiap kenaikan satu unit dalam `Account Age`, `Total Charges` meningkat sekitar 13.17 unit. Meskipun tren linier ini tampak cukup kuat, terdapat penyebaran yang cukup lebar pada akun dengan usia yang lebih lama, menunjukkan bahwa meskipun usia akun merupakan prediktor penting untuk total biaya, faktor lain mungkin juga berkontribusi terhadap variasi ini. Analisis lebih lanjut diperlukan untuk memahami variasi dalam total biaya yang tidak sepenuhnya dijelaskan oleh usia akun.

#### 2f. Analisislah distribusi ViewingHoursPerWeek dan identifikasi apakah ada pelanggan yang menonton secara signifikan lebih banyak atau lebih sedikit dibandingkan dengan mayoritas pelanggan lainnya.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualisasi distribusi ViewingHoursPerWeek menggunakan histogram dan KDE plot
plt.figure(figsize=(12, 6))
sns.histplot(df['ViewingHoursPerWeek'], bins=30, kde=True, color='blue')
plt.title('Distribusi Viewing Hours per Week')
plt.xlabel('Viewing Hours per Week')
plt.ylabel('Jumlah Pengguna')
plt.grid(True)
plt.show()

# Menggunakan boxplot untuk mengidentifikasi outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x='ViewingHoursPerWeek', data=df, color='orange')
plt.title('Boxplot Viewing Hours per Week')
plt.xlabel('Viewing Hours per Week')
plt.grid(True)
plt.show()

# Menghitung kuartil dan IQR untuk mengidentifikasi outliers
Q1 = df['ViewingHoursPerWeek'].quantile(0.25)
Q3 = df['ViewingHoursPerWeek'].quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas untuk outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identifikasi pelanggan yang menonton jauh lebih sedikit atau lebih banyak dari mayoritas
outliers_below = df[df['ViewingHoursPerWeek'] < lower_bound]
outliers_above = df[df['ViewingHoursPerWeek'] > upper_bound]

# Menampilkan analisis
print("\nKeterangan Analisis:")
print(f"1. Batas bawah untuk outliers (menonton jauh lebih sedikit): {lower_bound}")
print(f"2. Batas atas untuk outliers (menonton jauh lebih banyak): {upper_bound}")
print(f"3. Jumlah pelanggan yang menonton lebih sedikit dari batas bawah: {len(outliers_below)}")
print(f"4. Jumlah pelanggan yang menonton lebih banyak dari batas atas: {len(outliers_above)}")

print("\n5. Pelanggan yang menonton secara signifikan lebih sedikit mungkin tidak terlalu terlibat dalam layanan.")
print("   Ini dapat menjadi indikasi bahwa mereka lebih rentan untuk churn atau kurang tertarik dengan konten yang disediakan.")
print("\n6. Pelanggan yang menonton secara signifikan lebih banyak mungkin sangat terlibat atau pelanggan setia.")
print("   Mereka dapat menjadi target untuk peningkatan langganan atau fitur tambahan untuk meningkatkan pengalaman pengguna.")
```
